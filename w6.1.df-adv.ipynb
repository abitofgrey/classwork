```python
import pandas as pd

df = pd.read_csv("./data/french-ambs.csv")
df
```

```python
from random import choice

# import random
# random.choice
# this is overkill
# only import what you need

df["date_start"] = f"{df["year"]}-{choice(range(1,13)):02d}-{choice(range(1,29)):02d}" # datetime object: "yyyy-mm-dd"
df
```

```python
df = df.drop(columns=["date_start","lat_long"])
df = df[["name","place","year"]]
df
```

```python
# apply on data series

df["date_start"] = df["year"].apply(lambda x: f"{x}-{choice(range(1,13)):02d}-{choice(range(1,29)):02d}")
df
```

```python
import numpy as np

def get_date_start(row):
    rows = df.loc[
        (df["place"] == row["place"]) &
        (df["name"] == row["name"]) &
        (df["year"] == row["year"] - 1)
    ]
    if len(rows) > 0:
        return np.nan
    date_string = f"{row["year"]}-{choice(range(1,13)):02d}-{choice(range(1,29)):02d}"
    return date_string

df["date_start"] = df.apply(get_date_start, axis=1)
df
```

```python
def get_date_end(row):
    rows = df.loc[
        (df["place"] == row["place"]) &
        (df["name"] == row["name"]) &
        (df["year"] == row["year"] + 1)
    ]
    if len(rows) > 0:
        return np.nan
    date_string = f"{row["year"]}-{choice(range(1,13)):02d}-{choice(range(1,29)):02d}"
    return date_string

df["date_end"] = df.apply(get_date_end, axis=1)
df
```

```python
# sort values

df = df.sort_values(["place","name","year"], ascending=True)
df
```

```python
# bfill(), ffill()

df["date_start"] = df["date_start"].ffill()
df["date_end"] = df["date_end"].bfill()
df
```

```python
from datetime import datetime
fmt = "%Y-%m-%d"

df["date_end"] = df["date_end"].apply(lambda x: datetime.strptime(x, fmt))
df["date_start"] = df["date_start"].apply(lambda x: datetime.strptime(x, fmt))
df
```

```python
df["date_end"].apply(lambda x: x.year)
```

```python
placesdf = df.groupby(["place"]).count()
placesdf
```

```python
placesdf = df.groupby(["place"]).agg(
    {
    "name": "count",
    "year": "min",
    "year": "max",
    }
).reset_index()
placesdf
```

```python
placesdf = df.groupby(["place"]).agg(
    num=("year", "count"),
    min_year=("year", "min"),
    max_year=("year", "max"),
    ambs=("name", "unique"),
    num_ambs=("name", "nunique")
).reset_index().sort_values(["num"], ascending=True)
placesdf
```

```python
placesdf = df.groupby(["place","name"]).agg(
    num=("year", "count"),
    min_year=("year", "min"),
    max_year=("year", "max")
).reset_index().sort_values(["num"], ascending=False)
placesdf
```

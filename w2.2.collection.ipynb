```python
# Set
# mutable, unordered (no index), no duplicates, iterable

set_inferred = {1,2,3,2} # almost never do this
set_explicit = set([2,3,4,4])

print(f"set inferred = {set_inferred}")
print(f"set_explicit = {set_explicit}")
```

```python
set_inferred.add(7) # add
print(f"add = {set_inferred}")
set_inferred.remove(3) # remove
print(f"remove set inferred = {set_inferred}")
diff = set_inferred.difference(set_explicit) # get difference between sets
print(f"diff = {diff}")
overlap = set_inferred.intersection(set_explicit) # get overlaps between sets
print(f"overlap = {overlap}")
```

```python
# tupple
# immutable, ordered (index), duplicates, iterable

tuple_inferred = (1,2,3,3,5,6)
tuple_explicit = tuple([4,5,6,7,2])

print(f"tuple inferred = {tuple_inferred}")
print(f"tuple explicit = {tuple_explicit}")
```

```python
# index notation
# [start:stop:step]

print(f"index 0 = {tuple_inferred[0]}")
print(f"last item = {tuple_inferred[-1]}")
print(f"get 2nd and 3rd items = {tuple_inferred[1:3]}")
print(f"tuple inferred has {len(tuple_inferred)} elements")
```

```python
# tuple methods
count = tuple_inferred.count(3)
print(count)
idx = tuple_inferred.index(3)
print(idx)
```

```python
# list
# mutable, ordered (index), duplicates, iterable

list_inferred = [1,2,3,4,5,5,6,3]
list_explicit = list((4,5,6,1,1,3,2))

print(f"list inferred = {list_inferred}")
print(f"list explicit = {list_explicit}")
```

```python
# list methods

list_inferred.append(7)
print(f"append = {list_inferred}")
list_inferred.remove(5)
print(f"remove = {list_inferred}")
removed = list_inferred.pop()
print(f"removed = {removed}")
print(f"after pop = {list_inferred}")
```

```python
# changing

list_inferred[0] = "Awesome Index"
list_inferred
```

```python
# dictionary
# mutable, {key:value}, unordered {index by key}
# no duplicate keys, but duplicate values, iterable

dict_inferred = {
    "name": "John",
    "age": 30
}
dict_explicit = dict(city="new york", country="usa")

print(f"dict inferred = {dict_inferred}")
print(f"dict explicit = {dict_explicit}")
```

```python
# indexing by key

dict_inferred["age"]
```

```python
# add / change values

dict_inferred["gender"] = "m"
dict_inferred["age"] = 72
dict_inferred
```

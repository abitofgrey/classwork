# Variables

Variables are objects that store data

The name of a variable references a location in memory that stores the data specified when contructed

Every varible created is associated with a name, naming must follow these rules:

- Cannot begin with a number
- Cannot contain symbols
- Cannot contain spaces

There are also certain naming conventions that variables follow, but are not impossible

This is because it allows mutual understanding of code among different groups of people that may not be farmiliar with the program

For instance, classes usually use capitalization like "className", but variables usually use underscores and no capitals like "var_name"

## Simple Variables

- String (str)
- integer (int)
- float (float)
- boolean (bool)

## Methods / Math

```python
# Strings

string_simple = "This is a String."
string_function = str("Declaring a String")
string_int = "12345"
string_int2 = "67"

# Printing

print(string_simple)

# Variable Type Function

print(type(string_function))

# Concatination of Strings (Cannot Mathematically add Strings)

print(string_int + string_int2)
```

```python
# int

int_simple = 123
int_function = int(456)
int_from_string = int("789")

print(int_simple, type(int_simple))
print(int_function + int_from_string)
```

```python
# float

float_simple = 1.0
float_function = float(2)

print("float_simple + float_function = ", float_simple + float_function)
print(f"float_simple: {float_simple} + float_function: {float_function} = {float_simple + float_function}")
```

```python
print(f"coerce float to int: {int(9.9999999)}")
```

```python
# boolean

bool_simple = True
bool_function = bool(0)

print(bool_simple)
print(bool_function)
```

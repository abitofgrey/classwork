```python
# while loop

condition = True

while condition: # while this condition is true
    # run
    # this
    # code
    # block
    # over
    # and
    # over
    condition = False # until condition is false
```

```python
count = 0
while True:
    if count > 10:
        break
    print(count)
    count += 1
print("loop broken!")
```

```python
count = 0
while count <= 10:
    print(count)
    count += 1
print("loop done!")
```

```python
isAgeValid: bool = False

while not isAgeValid:
    age: str = input("Enter your age.")
    if not age.isnumeric():
        print("Number, please.")
        continue # continue to next iteration; skip rest of code block
    isAgeValid = True
    print(f"You are {age} years old.")

try:
    age = int(age)
except Exception as e:
    print(e)
```

```python
# for loop; iterates over collections
from typing import List
collection: List[int] = [1,2,3,4,5]

for num in collection:
    print(num)

# enumerate
# returns index and item in that index
# (idx, collection[idx])
for idx, num in enumerate(collection):
    print(f"{num} is at index {idx}")
```

```python
sentence: str = "This is a Python course"
words: List[str] = sentence.split(" ") # split returns a list
used_vowels: List[str] = []
vowel_options: List[str] = ["a", "e", "i", "o", "u", "y"]

for word in words:
    for letter in word:
        if letter in vowel_options:
            used_vowels.append(letter)

print(used_vowels)
```

```python
countries: List[str] = ["France", "Spain", "USA", "Argentina"]

countries_short: List[str] = [f"{country[:3]}." for country in countries]

print(countries_short)
```

```python
# functions
# takes input (arguments)
# processes it
# and return an output (return)

def shorten_names(names_list: List[str], chars: int = 3) -> List[str]:
    """
    Shortens the strings inside a list to chars characters.
    
    Args:
        names_list (List[str]): the list of string you want shortened to chars characters.
        chars (int): The number of characters to shorten each string in the list to; defaults to 3 characrters.
    
    Returns:
        List[str]: a new list of shortened strings appended with "."
    """
    # cannot read information from outside the function
    names_list = [f"{name[:chars]}." for name in names_list]
    return names_list

print(shorten_names(countries))
print(shorten_names(countries, chars = 2))
```

```python
shorten_names(["anme", "name", "mean"])
```

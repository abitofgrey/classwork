```python
# Conditionals:
# equal to ==
# greater than >
# greater than or equal to >=
# less than <
# less than or equal to <=
# not equal to !=

a: int = 10
b: int = 12

print(f"a = {a}, b = {b}")
print(f"a = b: {a==b}")
print(f"a > b: {a>b}")
print(f"a <= b: {a<=b}")
print(f"a != b: {a!=b}")
```

```python
# type hinting: str, int, float, bool
# in, or, and, is, not
from typing import List, Dict, Tuple, Set

c: List[int] = [1, 3, 10, 6, 7]

# in operator
print(f"is {a} in {c}?\n{a in c}")

# and operator
print(f"is {a} equal to {b} and is {a} in {c}?\n{a==b and a in c}")

# or operator
print(f"is {a} equal to {b} or is {a} in {c}?\n{a==b or a in c}")

# is operator
print(f"is {a} an int?\n{type(a) is int}")

# not operator
print(f"Not True: {not True}")
print(f"Not False: {not False}")
print(f"Not a < b: {not a < b}")
```

```python
# truthy / falsy

# falsy = 0, 0.0, "", None, [], {}, ()

print(bool(a))
print(bool(b))
print(bool(c))
print(bool(0))
print(bool([]))
```

```python
# if statements
# pass does nothing

condition: bool = False

if not a:
    print(f"condition is {a}")
elif 5 == "5":
    print("5")
else:
    print("made it through")
```

```python
# input

num: str = input("Enter a number b/w 1 and 10")

if num.isnumeric():
    num = int(num)
    if num >= 1 and num <= 10:
        print("yay for following directions")
    else:
        print("we asked for 1-10, bozo")
else:
    print("hey stupid, follow directions. i said a number")
```

```python
# try except
try:
    num: str = input("Enter a number b/w 1 and 10")
    if num.isnumeric():
        num = int(num)
        if num >= 1 and num <= 10:
            print("yay")
        else:
            raise Exception("follow directions, between 1 and 10")
    else:
        raise Exception("Choose a number")
except Exception as e:
    print(f"There was an error: {e}")
```

```python
num: str = input("Enter a number b/w 1 and 10 or 20 and 30")

try:
    if not num.isnumeric():
        raise Exception("Choose a number")
    num = int(num)
    if (num < 1 or num > 10) and (num < 20 or num > 30):
        raise Exception("Number should be between 1 and 10 or 20 and 30")
    
    print(f"{num} is following the rules")
except Exception as e:
    print(e)
```

```python
from scipy import stats
```

```python
pd_skew = df["height_in_cm"].skew()
scipy_skew = stats.skew(df["height_in_cm"], nan_policy = "omit")
print(f"pd_skew = {pd_skew}, scipy_skew = {scipy_skew}")
```

```python
pd_kurt = df["height_in_cm"].kurt()
scipy_kirt = stats.kurtosis(df["height_in_cm"].dropna())
scipy_kirt_omit = stats.kurtosis(df["height_in_cm"], nan_policy = "omit")
print(f"pd_kurt = {pd_kurt}, scipy_kurt = {scipy_kirt}, scipy_kurt_omit = {scipy_kirt_omit}")
```
